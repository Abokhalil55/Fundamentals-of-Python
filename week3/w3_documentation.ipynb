# 1. Identify the Components

## 1.1 Inputs
    Quiz 1 mark
    Quiz 2 mark
    Quiz 3 mark
    Continue choice: Y or N

## 1.2 Process
    Start with choice set to Y.
    While the user chooses to continue, ask for three quiz marks.
    Add the three marks and divide by 3 to calculate the average.
    Compare the average with 50.
    Display whether the student passed or failed.
    Ask whether another student's marks should be entered.

## 1.3 Output
    The student's average mark
    Pass message if the average is 50 or above
    Fail message if the average is below 50
    Program ended message when the user chooses not to continue

# 2. Design the Algorithm

## 2.1 Pseudocode
    Set choice to "Y"

    WHILE choice is "Y"
        INPUT quiz_1
        INPUT quiz_2
        INPUT quiz_3

        average = (quiz_1 + quiz_2 + quiz_3) / 3
        DISPLAY average

        IF average >= 50 THEN
            DISPLAY "Student passed the subject."
        ELSE
            DISPLAY "Student fail the subject."
        ENDIF

        INPUT choice
    ENDWHILE

    DISPLAY "Program Ended"

## 2.2 Algorithm Structure
### 2.2.1 Sequence
    The program follows ordered steps: get marks, calculate the average, display the result, then ask whether to continue.

### 2.2.2 Selection
    The if and else statements choose between the pass output and fail output based on the average mark.

### 2.2.3 Iteration
    The while loop allows the program to repeat for another student's marks while the user enters Y.
